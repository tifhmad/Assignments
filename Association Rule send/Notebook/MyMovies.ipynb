# Apriori Algorithm
## Association Rule with 0.10% support and confidence 0.70% interval

```python
frequent_itemsets=apriori(movie1,min_support=0.1,use_colnames=True)
frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.10)
rules
```

```python
rules[rules.lift>1]
```

```python
rules[['support','confidence']].hist()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.colorbar()
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

```python
frequent_itemsets=apriori(movie1,min_support=0.15,use_colnames=True)
frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x:len(x))
```

```python
rules1=association_rules(frequent_itemsets,metric='lift',min_threshold=0.10)
rules1[rules1.lift>1]
```

```python
rules1[['support','confidence']].hist()
```

```python
plt.scatter(rules1['support'],rules1['confidence'])
plt.colorbar()
plt.xlabel('Support')
plt.ylabel('Confidence')
```
