# Use Random Forest to prepare a model on fraud data 
treating those who have taxable_income <= 30000 as "Risky" and others are "Good"


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fraud=pd.read_csv('/content/Fraud_check (1).csv')
```

```python
fraud
```

```python
fraud=fraud.rename(columns={'Marital.Status':'marital_status','Taxable.Income':'taxable_income','City.Population':'city_population','Work.Experience':'wrok_experience'})
```

```python
fraud
```

```python
fraud.info()
```

```python
fraud.describe()
```

```python
def split(num):
  if num <30000:
    return 'Risky'
  else:
    return 'Good'
fraud.taxable_income=fraud.taxable_income.apply(split)
```

```python
fraud
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
fraud['Undergrad']=le.fit_transform(fraud['Undergrad'])
fraud['marital_status']=le.fit_transform(fraud['marital_status'])
fraud['Urban']=le.fit_transform(fraud['Urban'])
```

```python
fraud
```

```python
import seaborn as sns
sns.pairplot(fraud)
```

```python
corr=fraud.corr()
import seaborn as sns
sns.heatmap(corr,annot=True)
```

```python
x=fraud.iloc[:,fraud.columns!='taxable_income']
y=fraud['taxable_income']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=RandomForestClassifier(n_jobs=3,oob_score=True,n_estimators=15,criterion='entropy')
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
df=pd.DataFrame({'actual':y_test,'predicted':pred})
df
```

```python
print(classification_report(y_test,pred))
```
